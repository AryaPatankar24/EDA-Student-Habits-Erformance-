```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_habits_performance.csv')
```

#8.Is there a correlation between study_hours_per_day and exam_score?

```python
correlation = df['study_hours_per_day'].corr(df['exam_score'])
print(f"Correlation between Study Hours and Exam Score: {correlation:.2f}")
```

```python
sns.lmplot(x='study_hours_per_day', y='exam_score', data= df, line_kws= {'color': 'red'})
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study Hours per Day')
plt.ylabel('Exam Score')
plt.show()
```

#9.How does sleep_hours relate to mental_health_rating?

```python
correlation = df['sleep_hours'].corr(df['mental_health_rating'])
print(f"Correlation between sleep HOurs and Menal Health Rating: {correlation:.2f}")
```

```python
sns.lmplot(x='sleep_hours', y='mental_health_rating', data= df, line_kws={'color': 'blue'})
plt.title('Sleep Hours vs Mental Health Rating')
plt.xlabel('Sleep Hours')
plt.ylabel('Mental Health Rating')
plt.show()
```

#10.Compare exam_score across different diet_quality levels.

```python
sns.boxplot(x='diet_quality', y='exam_score', data=df, palette= 'Set3')
plt.title('Exam Score by Diet Quality')
plt.xlabel('Diet Quality')
plt.ylabel('Exam Score')
plt.show()
```

```python
grouped_scores = df.groupby('diet_quality')['exam_score'].mean().sort_values(ascending= False)
print(grouped_scores)
```

#11.Does gender affect study_hours_per_day or exam_score?

```python
# Boxplots for visual comparison
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Study hours by gender
sns.boxplot(x='gender', y='study_hours_per_day', data=df, ax=axes[0], palette='coolwarm')
axes[0].set_title('Study Hours per Day by Gender')
axes[0].set_xlabel('Gender')
axes[0].set_ylabel('Study Hours')

# Exam scores by gender
sns.boxplot(x='gender', y='exam_score', data=df, ax=axes[1], palette='coolwarm')
axes[1].set_title('Exam Score by Gender')
axes[1].set_xlabel('Gender')
axes[1].set_ylabel('Exam Score')

plt.tight_layout()
plt.show()

# Optional: Group-wise means
gender_stats = df.groupby('gender')[['study_hours_per_day', 'exam_score']].mean()
print(gender_stats)
```

#12.Do students with part-time jobs have lower attendance_percentage or exam_score?

```python
# Boxplots to compare attendance and exam scores by job status
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Attendance
sns.boxplot(x='part_time_job', y='attendance_percentage', data=df, palette='Set2', ax=axes[0])
axes[0].set_title('Attendance Percentage by Part-Time Job Status')
axes[0].set_xlabel('Has Part-Time Job')
axes[0].set_ylabel('Attendance Percentage')

# Exam Score
sns.boxplot(x='part_time_job', y='exam_score', data=df, palette='Set2', ax=axes[1])
axes[1].set_title('Exam Score by Part-Time Job Status')
axes[1].set_xlabel('Has Part-Time Job')
axes[1].set_ylabel('Exam Score')

plt.tight_layout()
plt.show()

# Optional: Compare averages
job_stats = df.groupby('part_time_job')[['attendance_percentage', 'exam_score']].mean()
print(job_stats)
```
