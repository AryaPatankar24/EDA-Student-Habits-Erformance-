```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_habits_performance.csv')
```

#16.Average exam_score by gender.

```python
avg_score = df.groupby('gender')['exam_score'].mean().sort_values(ascending =False)
print(avg_score)

sns.barplot(x=avg_score.index, y=avg_score.values, palette='pastel')
plt.title('Average Exam Score by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Exam Score')
plt.ylim(0, 100)
plt.show()
```

#17.Compare mental_health_rating between students with and without part-time jobs.

```python
sns.boxplot(x= 'part_time_job', y='mental_health_rating', data= df, palette= 'Set1')
plt.title('Mental Health Rating by Part-Time Job Status')
plt.xlabel('Has Part-Time Job')
plt.ylabel('Mental Health Rating')
plt.show()
```

```python
mh_means = df.groupby('part_time_job')['mental_health_rating'].mean()
print(mh_means)
```

#18.What is the average study_hours_per_day for students who participate in extracurriculars?

```python
study_avg = df.groupby('extracurricular_participation')['study_hours_per_day'].mean()
print(study_avg)
```

```python
sns.barplot(x= study_avg.index, y= study_avg.values, palette= 'cool')
plt.title('Average Study Hours/Day by Extracurricualr Participation')
plt.xlabel('Participates in Extracurricular')
plt.ylabel('Average Study Hours per Day')
plt.ylim(0, df['study_hours_per_day'].max() + 1)
plt.show()
```

#19.Which group scores better: students with good or poor internet quality?

```python
# Grouped average exam scores by internet quality
avg_scores_by_internet = df.groupby('internet_quality')['exam_score'].mean().sort_values(ascending=False)
print(avg_scores_by_internet)

# Bar plot
sns.barplot(x=avg_scores_by_internet.index, y=avg_scores_by_internet.values, palette='viridis')
plt.title('Average Exam Score by Internet Quality')
plt.xlabel('Internet Quality')
plt.ylabel('Average Exam Score')
plt.ylim(0, 100)  # Assuming score is out of 100
plt.show()
```
