```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student_habits_performance.csv')
```

#
1.	What is the average age of students?

```python
average_age = df['age'].mean()
print(f"AVerage Age of Students: {average_age:.2f} years")
```

```python
sns.histplot(df['age'], bins=10, kde=True)
plt.axvline(average_age, color='red', linestyle='--', label=f'Average Age: {average_age:.2f}')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Student Ages')
plt.legend()
plt.show()
```

#2.	What is the distribution of gender in the dataset?

```python
gender_counts = df['gender'].value_counts()
print(gender_counts)
```

```python
plt.figure(figsize=(8,6))
gender_counts.plot.pie(autopct='%1.1f%%', colors=['#66b3ff', '#ff9999', '#99ff99'])
plt.title('Gender Distribution of Students')
plt.ylabel('')
plt.show()
```

#3.What are the mean, median, and standard deviation for study_hours_per_day, sleep_hours, and exam_score?

```python
columns =['study_hours_per_day', 'sleep_hours', 'exam_score']
```

```python
summary_stats = df[columns].agg(['mean', 'median', 'std']).T
summary_stats.columns = ['Mean', 'Meadian', 'Standard Deviation']
print(summary_stats)
```

#4.What is the distribution of study_hours_per_day?

```python
mean_study = df['study_hours_per_day'].mean()
median_study = df['study_hours_per_day'].median()
```

```python
sns.histplot(df['study_hours_per_day'], bins= 10, kde= True, color= 'skyblue')
plt.axvline(mean_study, color='red', linestyle= '--', label= f'Mean: {mean_study:.2f}')
plt.axvline(median_study, color= 'green', linestyle= '-.', label= f'Median: {median_study:.2f}')
plt.title('Distribution of Study Hours per Day')
plt.xlabel('Study Hours per Day')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

#5.What is the distribution of netflix_hours and social_media_hours?

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
sns.histplot(df['netflix_hours'], bins= 10, kde= True, ax= axes[0], color= 'purple')
axes[0].set_title('Distribution of Netflix Hours per Day')
axes[0].set_xlabel('Netflix Hours')
axes[0].set_ylabel('Number of Students')
```

```python
sns.histplot(df['social_media_hours'], bins= 10, kde= True, ax= axes[1], color= 'orange')
axes[1].set_title('Distribution of Social Media Hours per Day')
axes[1].set_xlabel('Social Media Hours')
axes[1].set_ylabel('Number of Students')

plt.tight_layout()
plt.show()
```

#6.How are attendance_percentage values spread across students?

```python
mean_attendance = df['attendance_percentage'].mean()
median_attendance = df['attendance_percentage'].median()
```

```python
sns.histplot(df['attendance_percentage'], bins= 10, kde= True, color= 'teal')
plt.axvline(mean_attendance, color= 'red', linestyle= '--', label= f'Mean: {mean_attendance:.2f}%')
plt.axvline(median_attendance, color='green', linestyle= '--', label= f'Median: {median_attendance:.2f}%')
plt.title('Distribution of Attendance Percentage')
plt.xlabel('Attendance Percentage')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
sns.boxplot(x=df['attendance_percentage'], color='lightgreen')
plt.title('Boxplot of Attendance Percentage')
plt.xlabel('Attendance Percentage')
plt.show()
```

#7.Which parental education level is most common among students?

```python
parent_edu_counts = df['parental_education_level'].value_counts()
print(parent_edu_counts)
```

```python
sns.countplot(y='parental_education_level', data=df, order=parent_edu_counts.index, palette='Set2')
plt.title('Distribution of Parental Education Level')
plt.xlabel('Number of Students')
plt.ylabel('Parental Education Level')
plt.show()
```
